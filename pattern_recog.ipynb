```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set();
```

```python
#read data

food_prod = pd.read_excel(/africa_food_production.xlsx")
food_prod
```

```python
#check for NAN

food_prod.isnull().sum()
```

```python
food_prod.describe()
```

```python
#drop "Year" column. It is an integer and will not be required. dates for data is 2004 to 2013

food_prod.drop(["Year", "Kilo_ton_per_annum"], axis = 1, inplace = True)
```

```python
#aggregate for data

df = food_prod.groupby("Country").sum()
df
```

```python
#bar plot

df.plot(figsize = (20, 20), kind ='bar')
plt.locator_params("y", nbins = 10)
plt.title("African Countries Food Production per day 2004 to 2013", fontsize = 25)
plt.xticks(fontsize = 14, rotation = 90)
plt.yticks(fontsize = 14)
plt.legend(fontsize = 20)
plt.xlabel("Country", fontsize = 20)
plt.ylabel("Food production / day (Kilo Tons)", fontsize = 20);
```

```python
#aggregate for data

df1 = food_prod.groupby("Item").sum()
df1
```

```python
#bar plot

df1.plot(figsize = (20, 20), kind ='bar')
plt.locator_params("y", nbins = 30)
plt.title("Food Types Produced in Africa per day 2004 to 2013", fontsize = 25)
plt.xticks(fontsize = 12, rotation = 90)
plt.yticks(fontsize = 16)
plt.legend(fontsize = 20)
plt.xlabel("Food Type", fontsize = 20)
plt.ylabel("Food Type Production / day (Kilo Tons)", fontsize = 20);
```

```python
#read data

food_supply = pd.read_csv(/africa_food_supply.csv")
food_supply
```

```python
#check for NAN

food_supply.isnull().sum()
```

```python
food_supply.describe()
```

```python
#drop "Year" column. will not be required. our data spans from 2004 to 2013

food_supply.drop(["Year"], axis = 1, inplace = True)
```

```python
#aggregate for data

df2 = food_supply.groupby("Country").sum()
df2
```

```python
#bar plot

df2.plot(figsize = (20, 20), kind ='bar')
plt.title("Food Supplied in African Countries per day 2004 to 2013", fontsize = 25)
plt.xticks(fontsize = 16, rotation = 90)
plt.yticks(fontsize = 16)
plt.legend(fontsize = 20)
plt.xlabel("Country", fontsize = 20)
plt.ylabel("Quantity of Food Supplied(kcal) / person / day", fontsize = 20);
```

```python
#merge dataframes to compare

df3 = pd.merge(df, df2, on = "Country")
df3
```

```python
#regression plot

plt.figure(figsize = (20, 10))
plt.xlabel("Value_kcal_per_person_per_day", fontsize = 18)
plt.ylabel("Food", fontsize = 18)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
sns.regplot(x = "Value_kcal_per_person_per_day", y = "Kilo_ton_per_day", data = df3);
```

```python
#bar plot: Production vs Consumption

df3.plot(figsize = (20, 20), kind ='bar')
plt.title("Food Produced vs Food Supplied in African Countries per day 2004 to 2013", fontsize = 25)
plt.xticks(fontsize = 16, rotation = 90)
plt.yticks(fontsize = 16)
plt.legend(fontsize = 20)
plt.xlabel("Country", fontsize = 20)
plt.ylabel("Food", fontsize = 20);
```

```python
#read data

food_supply = pd.read_csv(/africa_food_supply.csv")
```

```python
#convert int to datetime "Year" column

food_supply["Year"] = pd.to_datetime(food_supply["Year"], format = "%Y")
food_supply
```

```python
#plot boxplot: date vs food supplied

plt.figure(figsize = (20 , 20))
sns.set(font_scale = 2)
sns.boxplot(x = "Value_kcal_per_person_per_day", y = "Year", data = food_supply, orient = "h");
```

```python
#create new dataframe

df4 = food_supply.loc[food_supply['Year'] == '2004-01-01']
df4
```

```python
# describe new dataframe

df4.describe()
```

```python
#Visualizing outliers for 2004

df4.plot.scatter(x = "Country", y = "Value_kcal_per_person_per_day", c = "DarkBlue", figsize = (20, 20), alpha=0.5)
plt.xticks(fontsize = 14, rotation = 90)
plt.yticks(fontsize = 14)
plt.xlabel("Country", fontsize=18)
plt.ylabel("Food Supplied", fontsize = 18)
plt.legend(["Value_kcal_per_person_per_day"]);
```
